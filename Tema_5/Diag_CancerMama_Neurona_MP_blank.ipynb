El plano de separación descrito anteriormente se obtuvo utilizando el método de árbol de múltiples superficies (MSM-T) [K. P. El programa lineal real utilizado para obtener el plano de separación en el espacio tridimensional es el que se describe en: [K. P. Bennett y O. L. Mangasarian: "Robust Linear Programming Discrimination of Two Linearly Inseparable Sets", Optimization Methods and Software 1, 1992, 23-34].

### 2.4. Implementación de una MPNeuron más avanzada

```python
import numpy as np
from sklearn.metrics import accuracy_score

class MPNeuron:
    def __init__(self):
        self.threshold = None

    def model(self, x):
          # input : [x1,x2,x3,.....,xn][0||1]
        z = sum(x)
        return (z >= self.threshold)

    def predict(self, X):
          # input : [[1,0,1,0],[0,1,0,1]]
        Y = []
        for x in X:
            result = self.model(x)
            Y.append(result)
        return np.array(Y)

    def fit(self, X, Y):
        accuracy = {}
        # Seleccionar un threshold entre el # de caracteristicas de entrada 
        for th in range(X.shape[1] + 1):
            self.threshold = th
            Y_pred = self.predict(X)
            accuracy[th] = accuracy_score(Y_pred, Y)

        #Seleccionar el threshold que de mejores resultados 
        self.threshold = max(accuracy, key=accuracy.get)
```

Seguimos teniendo un problema debido a que en nuestro conjunto de datos las características de entrada reciben valores continuos, sin embargo, nuestra MPNeuron solo procesa características de entrada con valor binario

```python
# Para transformar un valor a binario
import matplotlib.pyplot as plt
print(pd.cut([0.04,2,3.1,5,6,0.02,0.6],bins=2, labels=[0,1]))
plt.hist([0.04,2,4,5,6,0.02,0.6],bins=2)
plt.show()
```

```python
# Transformamos las caracteríticas de entrada a un valor binario
X_train_bin = X_train.apply(pd.cut, bins=2, labels=[1,0])
X_test_bin = X_test.apply(pd.cut, bins=2, labels=[1,0])

X_train_bin
```

```python
# Instanciamos el modelo MPNeuron
mp_neurona = MPNeuron()

# Encontramos el threshold óptimo
mp_neurona.fit(X_train_bin.to_numpy(), y_train)
```

```python
# Threshold óptimo seleccionado
mp_neurona.threshold
```

```python
# Realizamos predicciones para ejemplos nuevos que no se encuentran en el conjunto de datos de entrenamiento
Y_Pred = mp_neurona.predict(X_test_bin.to_numpy())
```

```python
Y_Pred
```

```python
# Calculamos la exactitud de nuestra predicción
accuracy_score(y_test, Y_Pred)
```

```python
# Calculamos la matriz de confusión
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, Y_Pred)
```
